```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
from transformers import pipeline

def predict_weather(start_date, days):
    # Load dataset (modify the path as needed)
    file_path = '/content/drive/MyDrive/nyc_weather_2000_2024_celsius.csv'
    df = pd.read_csv(file_path)

    # Convert DATE column to datetime format
    df['DATE'] = pd.to_datetime(df['DATE'])
    df.rename(columns={'DATE': 'ds', 'TAVG': 'y'}, inplace=True)

    # Train Prophet model
    model = Prophet()
    model.fit(df)

    # Generate future dates
    future_dates = pd.date_range(start=start_date, periods=days, freq='D')
    future_df = pd.DataFrame({'ds': future_dates})

    # Predict temperatures
    forecast = model.predict(future_df)

    # Extract relevant columns
    predictions_df = forecast[['ds', 'yhat']]
    predictions_df.rename(columns={'ds': 'DATE', 'yhat': 'Predicted_TAVG'}, inplace=True)

    return predictions_df

# Load LLM for user interaction
llm = pipeline("text2text-generation", model="google/flan-t5-small")

def chat_with_llm():
    while True:
        user_input = input("Ask about the weather forecast (or type 'exit' to quit): ")
        if user_input.lower() == 'exit':
            break

        # Extract date from user input (basic parsing for demo purposes)
        import re
        match = re.search(r'\d{4}-\d{2}-\d{2}', user_input)
        if match:
            date = match.group()
            predictions = predict_weather(date, 1)
            temp = predictions.iloc[0]['Predicted_TAVG']
            response = f"The predicted temperature on {date} is {temp:.2f}°C."
        else:
            response = "Please ask about a specific date in YYYY-MM-DD format."

        print("AI:", response)

# Start chatbot
chat_with_llm()
```
